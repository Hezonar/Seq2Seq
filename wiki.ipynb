```python
import wikipediaapi
import re
from googletrans import Translator
import os
import csv
import pandas
```

```python
wiki_wiki = wikipediaapi.Wikipedia('MyProjectName (merlin@example.com)', 'ru')

def get_wiki_data(title, lang='ru'):
    page_py = wiki_wiki.page(title)
    if page_py.exists():
        return page_py.text
    else:
        return None

# Пример использования
source_text = get_wiki_data("Python", lang='ru')
source_text
```

```python
all_sentences = []
```

```python
def is_normal_text(text):
    if not text.strip():  # Проверка на пустую строку
        return False

    words = text.strip().split()
    if len(words) == 1 and len(words[0]) <= 3:  # Проверка на 1 слово с меньше 3 буквами
        return False
    
    pattern = re.compile(r'^[a-zA-Zа-яА-ЯёЁ\s!:\.;,№]*$')
    return bool(pattern.match(text))
```

```python
splited_text = source_text.split('\n')
for abzac in splited_text:
    sentences = abzac.split(".")
    for sentence in sentences:
        if is_normal_text(sentence):
            all_sentences.append(sentence.strip())
```

```python
def translate_to_english(text):
    translator = Translator()
    translation = translator.translate(text, src='ru', dest='en')
    return translation.text
```

```python
def save_data_csv(all_sentences, save_to_csv='data.csv'):
    X = []  # Список для оригинальных предложений
    Y = []  # Список для переведенных предложений
    csv_file_path = save_to_csv
    
    # Проверяем наличие файла и создаем заголовки, если файл не существует
    if not os.path.isfile(csv_file_path):
        with open(csv_file_path, 'a', newline='', encoding='utf-8') as csvfile:
            csv_writer = csv.writer(csvfile)
            csv_writer.writerow(['original text', 'translated text'])
    
    # Переводим и записываем предложения
    for i, sentence in enumerate(all_sentences):
        translated_sentence = translate_to_english(sentence)
        X.append(sentence)
        Y.append(translated_sentence)
    
        # Записываем в CSV файл
        with open(csv_file_path, 'a', newline='', encoding='utf-8') as csvfile:
            csv_writer = csv.writer(csvfile)
            csv_writer.writerow([sentence, translated_sentence])

        if i % 10 == 0:
            print(f"Processed: {i}/{len(all_sentences)}")
    return X, Y
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.nn.utils.rnn import pack_padded_sequence, pad_sequence
from torch.utils.data import DataLoader
from torchtext.data.utils import get_tokenizer
from torchtext.vocab import build_vocab_from_iterator
import spacy

# Загрузка моделей Spacy для русского и английского языков
tokenizer_ru = get_tokenizer('spacy', language='ru_core_news_sm')
tokenizer_en = get_tokenizer('spacy', language='en_core_web_sm')

# Переменные X и Y
X = X
Y = Y

# Функция для создания списка токенов из итератора данных
def yield_tokens(data_iter, language, tokenizer):
    for example in data_iter:
        yield tokenizer(example[language])

# Создание итератора данных
train_iter = [{'src': x, 'trg': y} for x, y in zip(X, Y)]

# Построение словарей
vocab_src = build_vocab_from_iterator(yield_tokens(train_iter, 'src', tokenizer_ru), specials=["<unk>"])
vocab_trg = build_vocab_from_iterator(yield_tokens(train_iter, 'trg', tokenizer_en), specials=["<unk>", "<start>", "<end>"])

# Установка токенов по умолчанию
vocab_src.set_default_index(vocab_src["<unk>"])
vocab_trg.set_default_index(vocab_trg["<unk>"])

# Преобразование в индексы
X_indices = [torch.tensor([vocab_src[token] for token in tokenizer_ru(sentence)]) for sentence in X]
Y_indices = [torch.tensor([vocab_trg["<start>"]] + [vocab_trg[token] for token in tokenizer_en(sentence)] + [vocab_trg["<end>"]]) for sentence in Y]

# Применение pad_sequence
X_padded = pad_sequence(X_indices, padding_value=vocab_src["<pad>"])
Y_padded = pad_sequence(Y_indices, padding_value=vocab_trg["<pad>"])

# Определение модели Seq2Seq
class Seq2SeqModel(nn.Module):
    def __init__(self, input_size, output_size, hidden_size, batch_size):
        super(Seq2SeqModel, self).__init__()
        self.hidden_size = hidden_size
        self.batch_size = batch_size
        self.embedding_src = nn.Embedding(input_size, hidden_size)
        self.encoder = nn.LSTM(hidden_size, hidden_size, num_layers=1, bidirectional=False)

        self.embedding_trg = nn.Embedding(output_size, hidden_size)
        self.decoder = nn.LSTM(hidden_size, hidden_size, num_layers=1, bidirectional=False)
        self.fc_out = nn.Linear(hidden_size, output_size)
        self.batch_size = batch_size

    def forward(self, src, trg, src_lengths):
        embedded_src = self.embedding_src(src)
        packed_src = pack_padded_sequence(embedded_src, src_lengths, batch_first=True, enforce_sorted=False)
        output, (hidden, cell) = self.encoder(packed_src)

        # Используем скрытое состояние и ячейку после каждого временного шага декодера
        trg_input = trg[:-1, :]
        embedded_trg = self.embedding_trg(trg_input)

        # Инициализируем скрытое состояние и ячейку для декодера
        hidden_dec = hidden[:, :1, :].contiguous()  # Передаем только первый элемент для батча размером 1
        cell_dec = cell[:, :1, :].contiguous()

        # Передаем скрытое состояние и ячейку после каждого временного шага декодера
        outputs = []
        for time_step in range(embedded_trg.size(1)):
            output, (hidden_dec, cell_dec) = self.decoder(embedded_trg[:, time_step:time_step+1, :], (hidden_dec, cell_dec))
            outputs.append(output)

        # Объединяем результаты и применяем линейный слой
        outputs = torch.cat(outputs, dim=1)
        prediction = self.fc_out(outputs)

        return prediction

# Определение параметров модели
input_size = len(vocab_src)
output_size = len(vocab_trg)
hidden_size = 256
batch_size = 5  # Размер батча
print("Параметры: ", input_size, output_size, hidden_size, batch_size)

# Инициализация модели, оптимизатора и функции потерь
model = Seq2SeqModel(input_size, output_size, hidden_size, batch_size)
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()

# Создание DataLoader
train_loader = DataLoader(train_iter, batch_size=batch_size, shuffle=True)

# Обучение модели
num_epochs = 10

# Ваш цикл обучения
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    
    for batch in train_loader:
        # Преобразование в тензоры и упаковка последовательностей
        X_batch = pad_sequence([torch.tensor([vocab_src[token] for token in tokenizer_ru(sentence)]) for sentence in batch['src']], batch_first=True, padding_value=vocab_src["<pad>"])
        Y_batch = pad_sequence([torch.tensor([vocab_trg["<start>"]] + [vocab_trg[token] for token in tokenizer_en(sentence)] + [vocab_trg["<end>"]]) for sentence in batch['trg']], batch_first=True, padding_value=vocab_trg["<pad>"])
        
        X_lengths = torch.tensor([len(tokenizer_ru(sentence)) for sentence in batch['src']])
        
        # Получение предсказаний модели
        output = model(X_batch, Y_batch, X_lengths)

        # Преобразование выходов модели в одномерный формат
        output = output.view(-1, output.shape[2])

    ### DEBUG ###
        # Получение индексов предсказанных токенов
        data_lengths = torch.tensor([len(tokenizer_ru(sentence)) for sentence in data])
        _, predicted_indices = torch.max(output, 1)
    
        # Разделение предсказаний для каждого входного предложения
        predicted_tokens = []
        current_index = 0
        for length in data_lengths:
            predicted_tokens.append([vocab_trg.lookup_token(idx.item()) for idx in predicted_indices[current_index:current_index+length]])
            current_index += length

        print(predicted_tokens)
    ### DEBUG ###

        # Подготовка target с учетом pack_padded_sequence
        target = pack_padded_sequence(Y_batch[:, 1:].contiguous(), X_lengths, batch_first=True, enforce_sorted=False).data

    ### DEBUG ###
        # Получение индексов предсказанных токенов
        data_lengths = torch.tensor([len(tokenizer_ru(sentence)) for sentence in data])
        predicted_indices = target
    
        # Разделение предсказаний для каждого входного предложения
        predicted_tokens = []
        current_index = 0
        for length in data_lengths:
            predicted_tokens.append([vocab_trg.lookup_token(idx.item()) for idx in predicted_indices[current_index:current_index+length]])
            current_index += length

        print(predicted_tokens)
    ### DEBUG ###
        

        # Обрезка предсказаний и target до минимальной длины в батче
        loss = criterion(output[:target.size(0)], target)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()

    average_loss = total_loss / len(train_loader)
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {average_loss:.4f}')
```

```python
#### import torch
import torch.nn as nn
import torch.optim as optim
from torch.nn.utils.rnn import pad_sequence, pack_padded_sequence, pad_packed_sequence
from torchtext.data.utils import get_tokenizer
from torchtext.vocab import build_vocab_from_iterator
from torch.utils.data import DataLoader

class Seq2SeqModel(nn.Module):
    def __init__(self, input_dim, emb_dim, hidden_dim, output_dim, pad_idx):
        super().__init__()

        self.embedding = nn.Embedding(input_dim, emb_dim)
        self.encoder = nn.GRU(emb_dim, hidden_dim)
        self.decoder = nn.GRU(emb_dim, hidden_dim)
        self.fc_out = nn.Linear(hidden_dim, output_dim)
        self.pad_idx = pad_idx

    def forward(self, src, trg, src_lengths, trg_lengths):
        src_embedded = self.embedding(src)
        trg_embedded = self.embedding(trg)

        packed_src = pack_padded_sequence(src_embedded, src_lengths, batch_first=True, enforce_sorted=False)
        encoder_output, encoder_hidden = self.encoder(packed_src)

        # Repeat the last hidden state of the encoder for each time step in the decoder
        encoder_hidden = encoder_hidden.repeat(1, trg.size(1), 1)

        packed_trg = pack_padded_sequence(trg_embedded, trg_lengths, batch_first=True, enforce_sorted=False)
        decoder_output, _ = self.decoder(packed_trg, encoder_hidden)

        output, _ = pad_packed_sequence(decoder_output, batch_first=True)
        output = self.fc_out(output)
        return output

    def predict(self, src, src_lengths, max_len=50):
        src_embedded = self.embedding(src)
    
        packed_src = pack_padded_sequence(src_embedded, src_lengths, batch_first=True, enforce_sorted=False)
        encoder_output, encoder_hidden = self.encoder(packed_src)
    
        # Используем unsqueeze, чтобы добавить размерность батча в decoder_hidden
        decoder_hidden = encoder_hidden.new_zeros(self.decoder.num_layers, encoder_hidden.shape[1], encoder_hidden.shape[2])
    
        # Инициализируем целевую последовательность как пустую
        trg = torch.tensor([[vocab_trg["<start>"]]])  # Начальный символ последовательности
    
        # Генерация предсказания
        output_sequence = []
    
        for i in range(max_len):
            trg_embedded = self.embedding(trg)
            decoder_output, decoder_hidden = self.decoder(trg_embedded, decoder_hidden)
            output = self.fc_out(decoder_output)
            prediction = output.argmax(2)[:, -1].item()  # Получаем индекс символа с наибольшей вероятностью
            output_sequence.append(prediction)
            trg = torch.cat((trg, torch.tensor([[prediction]])), dim=1)
    
            if prediction == vocab_trg["<end>"]:  # Если достигнут символ конца последовательности
                break
    
        return output_sequence



class Seq2SeqTrainer:
    def __init__(self, model, optimizer, criterion, clip):
        self.model = model
        self.optimizer = optimizer
        self.criterion = criterion
        self.clip = clip

    def train_step(self, src, trg, src_lengths, trg_lengths):
        self.model.train()
        self.optimizer.zero_grad()
        output = self.model(src, trg, src_lengths, trg_lengths)

        output_dim = output.shape[-1]
        output = output[1:].view(-1, output_dim)
        trg = trg[1:].reshape(-1)

        loss = self.criterion(output, trg)
        loss.backward()

        torch.nn.utils.clip_grad_norm_(self.model.parameters(), self.clip)
        self.optimizer.step()

        return loss.item()

def prepare_data(X, Y, vocab_src, vocab_trg, tokenizer_ru, tokenizer_en):
    X_indices = [torch.tensor([vocab_src[token] for token in tokenizer_ru(sentence)]) for sentence in X]
    Y_indices = [torch.tensor([vocab_trg["<start>"]] + [vocab_trg[token] for token in tokenizer_en(sentence)] + [vocab_trg["<end>"]]) for sentence in Y]

    X_lengths = torch.tensor([len(sentence) for sentence in X_indices])
    Y_lengths = torch.tensor([len(sentence) for sentence in Y_indices])

    X_padded = pad_sequence(X_indices, padding_value=vocab_src["<pad>"]).T.long()
    Y_padded = pad_sequence(Y_indices, padding_value=vocab_trg["<pad>"]).T.long()

    return X_padded, Y_padded, X_lengths, Y_lengths

# Исходные предложения
X = X
Y = Y

# Загрузка моделей Spacy для русского и английского языков
tokenizer_ru = get_tokenizer('spacy', language='ru_core_news_sm')
tokenizer_en = get_tokenizer('spacy', language='en_core_web_sm')

# Создание итератора данных
train_dataset = [{'src': x, 'trg': y} for x, y in zip(X, Y)]

# Построение словарей
def yield_tokens(data_iter, language, tokenizer):
    for example in data_iter:
        yield tokenizer(example[language])

vocab_src = build_vocab_from_iterator(yield_tokens(train_dataset, 'src', tokenizer_ru), specials=["<unk>", "<pad>"])
vocab_trg = build_vocab_from_iterator(yield_tokens(train_dataset, 'trg', tokenizer_en), specials=["<unk>", "<pad>", "<start>", "<end>"])

vocab_src.set_default_index(vocab_src["<pad>"])
vocab_trg.set_default_index(vocab_trg["<pad>"])

all_tokens = list(yield_tokens(train_dataset, 'src', tokenizer_ru)) + list(yield_tokens(train_dataset, 'trg', tokenizer_en))

joint_vocab = build_vocab_from_iterator(
    all_tokens,
    specials=["<unk>", "<pad>", "<start>", "<end>"]
)

joint_vocab.set_default_index(joint_vocab["<pad>"])

vocab_src = joint_vocab
vocab_trg = joint_vocab

input_dim = len(vocab_src)
output_dim = len(vocab_trg)
emb_dim = 256
hidden_dim = 512
pad_idx = vocab_src["<pad>"]

model = Seq2SeqModel(input_dim, emb_dim, hidden_dim, output_dim, pad_idx)
criterion = nn.CrossEntropyLoss(ignore_index=pad_idx)
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Используем DataLoader для формирования батчей
BATCH_SIZE = 50
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)

trainer = Seq2SeqTrainer(model, optimizer, criterion, clip=1)
```

```python
def prepare_data_inference(sentence, vocab, tokenizer):
    indices = torch.tensor([vocab[token] for token in tokenizer(sentence)])
    lengths = torch.tensor([len(indices)])

    padded = pad_sequence([indices], padding_value=vocab["<pad>"]).T.long()

    return padded, lengths

def infer(model, sentence, vocab_src, vocab_trg, tokenizer_ru, tokenizer_en, max_length=50):
    model.eval()
    with torch.no_grad():
        X_padded, X_lengths = prepare_data_inference(sentence, vocab_src, tokenizer_ru)

        # Заготовка для предсказанного перевода
        trg_init_token = vocab_trg["<start>"]
        trg_token = torch.tensor([trg_init_token])

        # Передача данных через модель
        output = model(X_padded, trg_token.unsqueeze(0), X_lengths, torch.tensor([1]))

        # Получение индекса наилучшего предсказания
        pred_token = output.argmax(2)[:, -1].item()

        # Постепенное предсказание токенов, пока не встретится токен завершения
        count = 0
        while pred_token != vocab_trg["<end>"] and count < max_length:
            trg_token = torch.cat((trg_token, torch.tensor([pred_token])))
            output = model(X_padded, trg_token.unsqueeze(0), X_lengths, torch.tensor([trg_token.size(0)]))
            pred_token = output.argmax(2)[:, -1].item()
            count += 1

        # Индексы предсказанных токенов (исключая <start> и <end>)
        pred_indices = trg_token[1:-1].tolist()

        # Восстановление предложения из индексов
        pred_sentence = ' '.join([vocab_trg.itos[idx] for idx in pred_indices if idx != vocab_trg["<start>"]])

        return pred_sentence

# Пример использования
for sentence in X:
    prediction = infer(model, sentence, vocab_src, vocab_trg, tokenizer_ru, tokenizer_en)
    print(f"Input: {sentence}")
    print(f"Prediction: {prediction}\n")
```

```python
def predict(model, data, vocab_src, vocab_trg, tokenizer_ru, tokenizer_en):
    # Предобработка данных
    data_indices = [torch.tensor([vocab_src[token] for token in tokenizer_ru(sentence)]) for sentence in data]
    data_padded = pad_sequence(data_indices, batch_first=True, padding_value=vocab_src["<pad>"])
    data_lengths = torch.tensor([len(tokenizer_ru(sentence)) for sentence in data])

    # Предсказание модели
    model.eval()  # Переключение в режим оценки
    with torch.no_grad():
        output = model(data_padded, data_padded, data_lengths)

    # Преобразование выходов модели в одномерный формат
    output = output.view(-1, output.shape[2])

    # Получение индексов предсказанных токенов
    _, predicted_indices = torch.max(output, 1)

    # Разделение предсказаний для каждого входного предложения
    predicted_tokens = []
    current_index = 0
    for length in data_lengths:
        predicted_tokens.append([vocab_trg.lookup_token(idx.item()) for idx in predicted_indices[current_index:current_index+length]])
        current_index += length

    return predicted_tokens

# Пример использования функции predict
data_to_predict = ["структура", "вычисления"]
predicted_tokens = predict(model, data_to_predict, vocab_src, vocab_trg, tokenizer_ru, tokenizer_en)
print(predicted_tokens)
```
